Reading an image into an array and displaying it 

```
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```
img = cv.imread('CH.jpg', 0)  
plt.imshow(img, cmap = 'gray')
plt.show()
```

Here: 

*   '0' stands for grayscale image
*   '1' stands for color  
*   '-1' stands for unchanged







```
plt.imshow(img, cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.show()
```

Ticks are the values used to show specific points on the coordinate axis.

If we Do not want to display the X and Y coordinates we can use 'plt.xticks([])' and plt.ytick([])

```
img = cv.imread('CH.jpg', 1)
plt.imshow(img, cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.show()
```

in CV the default order of the image is BGR instead of RGB which is expected by matplotlib, therefore we need to convert BGR image to RGB.  

```
img = cv.imread('CH.jpg', 1)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)   
plt.imshow(img, cmap = 'gray')
plt.xticks([]), plt.yticks([])
plt.show()
```
