```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```
img = cv2.imread('CH.jpg')
height, width = img.shape[:2] 

rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2), 90, .5)
rotated_image = cv2.warpAffine(img, rotation_matrix, (width, height))

plt.imshow(rotated_image)
```

We have divided the height and width of the image by 2 to rotate the image about the centre 

```
img = cv2.imread('CH.jpg')
height, width = img.shape[:2] 

rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2), 180, .5)
rotated_image = cv2.warpAffine(img, rotation_matrix, (width, height))

plt.imshow(rotated_image)
```

```
img = cv2.imread('CH.jpg')
height, width = img.shape[:2] 

rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2), 270, .5)
rotated_image = cv2.warpAffine(img, rotation_matrix, (width, height))

plt.imshow(rotated_image)
```
